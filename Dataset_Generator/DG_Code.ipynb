```python
import openai
import gradio as gr
import os
import pandas as pd
openai.api_type = "azure"
# Add Endpoints here.
openai.api_base = #"https://<<change will be here >>.openai.azure.com/"

openai.api_version = "2023-03-15-preview"
# Add Key here.
openai.api_key = #<<Enter Open AI API key.>>

# Here prompt engineering is done by defining - Role + Expected Task + Steps to consider before generating Data.
content = '''
You are a healthcare insurance technical assistant with 10+ years of experience in this space. You provide deep 
technical expertise and guidance to insurance companies, healthcare providers, and other stakeholders 
in the industry. You work closely with the company's IT department to develop and manage technology solutions 
that support the company's operations and customer service. Primary objective is to generate mockup data for 
healthcare insurance consultants which is extremely relevant and real. Apart from this task you should not answer 
any questions while softly saying no such questions. Before generating mockup data always 
make sure to ask necessary information to generate the data. The data should be generated in table format 
with comma separated style. If you are unable to generate the data respond the question, respond by saying 
\"I am unable to generate the data."
'''
# Prior reference is needed for conversational chat. This is stored in message dictionary list.
messages = [
    {"role": "system", "content": content},
]

# Chatbot Function     
def chatbot(input):
    if input:
        messages.append({"role": "user", "content": input})
        chat = openai.ChatCompletion.create(engine="<<Model Name>>",messages = messages,
                                                temperature=1,max_tokens=400,top_p=0.95,
                                                frequency_penalty=0,presence_penalty=0,stop=None)
        reply = chat.choices[0].message.content
        messages.append({"role": "assistant", "content": reply})
        return reply

# Simple intutive UI to be generated by gradio lib. Input and Output view defined.
inputs = gr.inputs.Textbox(lines=7, label="Chat with Us")
outputs = gr.outputs.Textbox(label="Reply")

# Interactive function of chatbot.
gr.Interface(fn=chatbot, inputs=inputs, outputs=outputs, title="HealthCare Data Generator",
             description="Genarate mock Data",theme="compact").launch(share=True)

# This will generate local and public URL. You can share public URL for audience to access from anywhere 
# and explore.
```
